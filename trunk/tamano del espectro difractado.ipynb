```python
def espectroAngular2(f,dx,dy,z,Lambda):

	M = f.shape
	[u,v] = mgrid[-M[0]/2:M[0]/2,-M[1]/2:M[1]/2]
	du = 1./(M[0]*dx)
	dv = 1./(M[1]*dy)
	gaux = 1 - (Lambda*u*du)**2 - (Lambda*v*dv)**2
	gaux = where(gaux<=1, gaux, 0)
	k = 2*pi/Lambda
	G = exp( 1.j*z*k*sqrt( gaux ) )
	f1 = ifft2( fft2(f)*fftshift(G) )
	return f1,dx,dy
```

```python
def fresnel2(f,dx,dy,z,Lambda):
	M = f.shape
	[x,y] = mgrid[-M[0]/2:M[0]/2,-M[1]/2:M[1]/2]
	k=2*pi/Lambda
	dx2 = Lambda*z/M[0]/dx
	dy2 = Lambda*z/M[1]/dy
	w1 = exp(1.j*k/2/z*( (x*dx2)**2 + (y*dy2)**2 ))
	w2 = exp(1.j*k/2/z*( (x*dx )**2 + (y*dy )**2 ))
	f1 = fftshift( w1*fft2( fftshift(f*w2)))
	return f1,dx2,dy2
```

```python
def espectroAngular1(f,dx,z,Lambda):
	M =len(f)
	u = arange(-M/2,M/2)
	du = 1./(M*dx)
	gaux = 1 - (Lambda*u*du)**2 
	gaux = where(gaux<=1, gaux, 0)
	k = 2*pi/Lambda
	G = exp( 1.j*z*k*sqrt( gaux ) )
	f1 = ifft( fft(f)*fftshift(G) )
	return f1,dx
```

```python
def fresnel1(f,dx,z,Lambda):
	M =len(f)
	x = arange(-M/2,M/2)  
	k=2*pi/Lambda
	dx2 = Lambda*z/M/dx
	w1 = exp(1.j*k/2./z*( (x*dx2)**2 ))
	w2 = exp(1.j*k/2./z*( (x*dx )**2 ))
	f1 = ifftshift( w1*ifft( fftshift(f*w2)))
	return f1,dx2	
```

```python
### dimensiones de los pixeles del modulador
M1=44.603
M2=44.738
R1=8./M1
R2=8./M2
R=(R1+R2)*.5
```

```python
############ Parametros
M=10000 ### Tamano de la matriz aumentada
d0=R #um
Lambda=.550
############

u=zeros(M)
T=5.
d=T/R
u[M/2-d:M/2+d]=1
figsize(25,2)

s=arange(-M/2,M/2)*d0
plot(s,u)
```

```python
#rec,d=espectroAngular1(u,T,250000,Lambda)
rec,d=fresnel1(u,R,250000,Lambda)
figsize(25,2)
#xlim(0,M*d0)
s=arange(-M/2,M/2)*d/1000.
plot(s,abs(rec))
```
